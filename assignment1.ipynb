```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Also we can see the head of our csv
df = pd.read_csv("/content/cars_2021_raw.csv")
df.head()
```

**Now we use the describe method to check for missing values or data types.**

```python
df.describe()
```

**We observe that there is most of the needed information, of int or float types, yet we cannot see the price values**

```python
print(df['PRECIO'].dtype)
```

**Above we can see that the values of price are of object type, so we need to convert them to float type. For that we also need to delete the euro sign and change all the ',' to '.'.**

```python
df['PRECIO'] = df['PRECIO'].str.replace('€', '').str.replace(',', '')
df['PRECIO'] = df['PRECIO'].astype(float)
```

**Now if we run the describe method again we can see that the price values are now visible.**

```python
df.describe()
```

**Now we will check for missing values.**

```python
df.info()
```

```python
df.isnull().sum()
```

**From the above results we can deduce that the missing values are MNAR as exactly 39 cars have exactly same missing columns.**

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing Values Before Cleaning')
plt.show()
```

**The above heatmap shows where the missing values are present.**

```python
missing_rows = df[df['Tracción'].isnull()]
print(missing_rows['COMBUSTIBLE'].value_counts())
print(missing_rows['BRAND'].value_counts())
```

**As we can see most of the cars are Diesel cars, which is unexpected as diesel cars usually include specs like consumption per 100km, which is missing in this case. Also, the majority of the cars are of the brand Nissan. We concluded that these are a systematic error by Nissan. So, we decided to drop these rows as we believe they are most likely corrupted data and only represent 0.73% of the data.**

```python
df_clean = df.dropna(subset=['Tracción'])  # We can drop those rows using this method as all the cars share corrupted data columns
```

```python
sns.heatmap(df_clean.isnull(), cbar=False)
plt.title('Missing Values After Cleaning')
plt.show()
```

**Now we're going to identify the rows with 0 velocity and grah the distribution so we can understand the effect of velocity = 0 on the whole dataset to make an educated choice **

```python
zero_velocity_count = df[df['Velocidad Máxima (km/h)'] == 0].shape[0]
print(f"Number of cars with 0 velocity: {zero_velocity_count}")
```

```python
sns.histplot(df['Velocidad Máxima (km/h)'], bins=30, kde=True)
plt.title("Velocity Distribution Before Handling 0 Values")
plt.xlabel("Velocity")
plt.ylabel("Frequency")
plt.show()
```

From this diagram we can see that there is a significant frequency of cars where velocity is 0 given also the fact that the number of cars with 0 velocity are 495. These rows can be considered as outliers as it was also shown in the boxplot, but they can also be considered invalid entries.
Possibilities:
- remove rows: these entries can be considered outliers or invalid.
- replace with valid velocity: replace using the mean/median/mode velocity in the dataset
- cars are not working anymore: 0 velocity might imply that these cars are not working anymore. In this case the age and model of the car should be considered as older cars tend to be non-working

```python
print(df[df['Velocidad Máxima (km/h)'] == 0].describe())
```

this output suggests that the cars are not necessarely non-working cars as Motor cc, fuel tank, autonomia (km) outputs suggest that the cars are capable of working and being functional.
However, accelleration on average is 1.35 seconds which is low, which might suggest to data incosistecy.
